```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('/content/data/gapminder.csv')
```

```python
os.makedirs('images', exist_ok=True)

# create histogram
plt.figure(figsize=(8,5))
plt.hist(df['lifeExp'], edgecolor='black', alpha=0.7)

# labels and title
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.title('Histogram of Life Expectancy')

# save the figure
plt.savefig('images/life_exp_hist.png')

plt.show()
```

```python
# create histogram
plt.figure(figsize=(8,5))
plt.hist(df['gdpPercap'], edgecolor='black', alpha=0.7)

# labels and title
plt.xlabel('GDP')
plt.ylabel('Frequency')
plt.title('Histogram of GDP')

# save the figure
plt.savefig('images/gdp_hist.png')

plt.show()
```

```python
# create histogram with required settings
plt.figure(figsize=(8,5))
plt.hist(
    df['lifeExp'],
    bins=50,
    edgecolor='black',
    alpha=0.5,
    color='lightblue'
)

# same labels as problem 5
plt.xlabel('Life Expectancy')
plt.ylabel('Frequency')
plt.title('Histogram of Life Expectancy')

# save the figure
plt.savefig('images/life_exp_hist.png')

# add grid
plt.grid()

plt.show()
```

```python
plt.figure(figsize=(8,5))

plt.scatter(np.log(df['gdpPercap']), df['lifeExp'])

plt.xlabel('GDP Per Capita')
plt.ylabel('Life Expectancy (years)')
plt.title('GDP vs. Life Expectancy')

plt.show()
```

```python
# create histogram using pandas
plt.figure(figsize=(8,5))
df['gdpPercap'].hist()

plt.xlabel('GDP Per Capita')
plt.ylabel('Frequency')
plt.title('Histogram of GDP Per Capita')

# save to images folder
plt.savefig('images/gdp_hist_pandas.png')

plt.show()
```
